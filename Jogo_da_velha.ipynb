```python
import os
import random

jogarDnovo = "s"
jogadas = 0
quemJoga = 2
maxJogadas = 9
vit = "n"
velha = [
  [" "," "," "],
  [" "," "," "],
  [" "," "," "]
]

def tela():
      global velha
      os.system('clear')
      print("    0    1   2 ")
      print("0:   " + velha[0][0] + " | " + velha[0][1] + " | " + velha[0][2])
      print("    -----------")
      print("1:   " + velha[1][0] + " | " + velha[1][1] + " | " + velha[1][2])
      print("    -----------")
      print("2:   " + velha[2][0] + " | " + velha[2][1] + " | " + velha[2][2])
      print("Jogadas " + str(jogadas))

def jogadorJoga():
    global jogadas
    global quemJoga
    global maxJogadas
    if quemJoga == 2 and jogadas<maxJogadas:
        try: #verificador
            l= int(input("Linha: "))
            c= int(input("Coluna: "))
            while velha[l][c]!= " ":
                l= int(input("Linha: "))
                c= int(input("Coluna: "))
            velha[l][c]="X"
            jogadas+=1
            quemJoga=1
        except:
            print("Linha ou coluna invalida")

def cpuJoga():
    global jogadas
    global quemJoga
    global maxJogadas
    if quemJoga == 1 and jogadas<maxJogadas:
        l= random.randrange(0,3)
        c= random.randrange(0,3)
        while velha[l][c]!= " ":
            l= random.randrange(0,3)
            c= random.randrange(0,3)
        velha[l][c]="O"
        jogadas+=1
        quemJoga=2

def verificarVit():
    global velha
    vitoria="n"
    simbolos=["X","O"]
    for s in simbolos:
        vitoria = "n"
        #verifica linhas
        il=ic=0
        while il<3:
          soma=0
          ic=0
          while ic<3:
            if(velha[il][ic]==s):
                soma+=1
            ic+=1
          if(soma==3):
              vitoria=s
              break
          il+=1
        if(vitoria!="n"):
            break
        
        #verifica colunas
        il=ic=0
        while ic<3:
          soma=0
          il=0
          while il<3:
            if(velha[il][ic]==s):
              soma+=1
            il+=1
          if(soma==3):
              vitoria=s
              break
          ic+=1
        if(vitoria!="n"):
            break
        
        #verifica diagonal1
        soma=0
        iDiagonal=0
        while iDiagonal<3:
          if(velha[iDiagonal][iDiagonal]==s):
              soma+=1
          iDiagonal+=1
        if(soma==3):
              vitoria=s
              break

        #verifica diagonal2
        soma=0
        iDiagonal=0
        iDiagonac=2
        while iDiagonac>=0:
          if(velha[iDiagonal][iDiagonac]==s):
              soma+=1
          iDiagonal+=1
          iDiagonac-=1
        if(soma==3):
              vitoria=s
              break
    return vitoria

def redefinir():
    global jogadas
    global quemJoga
    global maxJogadas
    global vit
    global velha
    jogadas = 0
    quemJoga = 2
    maxJogadas = 9
    vit = "n"
    velha = [
  [" "," "," "],
  [" "," "," "],
  [" "," "," "]
]

while(jogarDnovo=="s"):
      while True:
        tela()
        jogadorJoga()
        cpuJoga()
        tela()
        vit=verificarVit()
        if(vit!="n")or(jogadas>=maxJogadas):
            break

      print("Fim de jogo")
      if(vit=="X" or vit=="O"):
          print("Resultado: Jogador " +vit+ " ganhou")
      else:
        print("Resultado: Empate")
      jogarDnovo=input("Jogar de Novo? [s/n]: ")
      redefinir()
```
