````python

print('Seja bem ao nosso pejeto!')
print('Abaixo ser√° exibido um menu contendo as moedas que s√£o poss√≠vel \nser observado seu valor em relac√£o ao real brasileiro')
print("\n -- Menu -- ")
print('1  - D√≥lar Americano (USD)')
print('2  - Euro (EUR)')
print('3  - Iene Japon√™s (JPY)')
print('4  - Libra Esterlina (GBP)')
print('5  - Franco Su√≠√ßo (CHF)')
print('6  - D√≥lar Canadense (CAD)')
print('7  - D√≥lar Australiano (AUD)')
print('8  - Coroa Norueguesa (NOK)')
print('9  - Coroa Dinamarquesa (DKK)')
print('10 - Renminbi Chin√™s (CNY)')
print(' -- ---- -- \n')
opcao = int(input('Qual moeda dejesa verificar seu valor? Digite apenas o n√∫mero '))
coin = ''

if opcao == 1:
  coin = 'USD'
elif opcao == 2:
  coin = 'EUR'
elif opcao == 3:
  coin = 'JPY'
elif opcao == 4: 
  coin = 'GBP'
elif opcao == 5:
  coin = 'CHF'
elif opcao == 6:
  coin = 'CAD'
elif opcao == 7: 
  coin = 'AUD'
elif opcao == 8:
  coin = 'NOK'
elif opcao == 9:
  coin = 'DKK'
elif opcao == 10:
  coin = 'CNY'


from bs4 import BeautifulSoup
import requests

Part_I = 'https://www.google.com/finance/quote/'
Part_F = '-BRL'

pagina = Part_I + coin + Part_F

page = requests.get(pagina) #USD-BRL, EUR-BLR, CNY-BRL -> √© s√≥ trocar depois do √∫timo /
soup = BeautifulSoup(page.text, 'html.parser')

list_div = soup.find(class_='YMlKec fxKbKc')

valuecoin = 0.0
for palavra in list_div:
  valuecoin = float(palavra)

valuecoin = "{:.2f}".format(valuecoin)

print('\nO valor da {0} no real brasileiro nesse exato momento: \n'.format(coin))

print('1 {0} ‚âÖ R$ {1}'.format(coin, valuecoin))
````


---


````python
import matplotlib.pyplot as plt #importando a fun√ß√£o de grafico da Biblioteca matplotlib 

listcoin_names = ['USD', 'EUR', 'JPY', 'GBP', 'CHF', 'CAD', 'AUD', 'NOK', 'DKK', 'CNY']
listcoin_values = []

for namecoin in listcoin_names: 
  pagina2 = Part_I + namecoin + Part_F
  page2 = requests.get(pagina2)
  soup2 = BeautifulSoup(page2.text, 'html.parser')
  list_div2 = soup2.find(class_='YMlKec fxKbKc')

  for V in list_div2:
    listcoin_values.append(float('{:.2f}'.format(float(V))))

#Gerando Grafico
plt.bar(listcoin_names, listcoin_values)
plt.yticks(listcoin_values)
plt.title('Valor das moedas no real brasileiro')
plt.ylabel('Valor da moeda - Em R$')
plt.xlabel('Moedas')
plt.show()
````
